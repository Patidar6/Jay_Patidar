```python
from rdkit.Chem import AllChem, MACCSkeys, rdMolDescriptors
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.ML.Descriptors import MoleculeDescriptors
import pandas as pd
```

```python
##Calculate descriptors using RDkit
def RDkit_descriptors(smiles):
    mols = [Chem.MolFromSmiles(i) for i in smiles] 
    calc = MoleculeDescriptors.MolecularDescriptorCalculator([x[0] for x in Descriptors._descList])
    desc_names = calc.GetDescriptorNames()
    
    Mol_descriptors =[]
    for mol in mols:
        # add hydrogens to molecules
        mol=Chem.AddHs(mol)
        # Calculate all 200 descriptors for each molecule
        descriptors = calc.CalcDescriptors(mol)
        Mol_descriptors.append(descriptors)
    return Mol_descriptors,desc_names 
```

### Morgan Fingerprinting

```python
###CALCULATE FP MORGAN 
def morgan_fpts(data):
    Morgan_fpts = []
    for i in data:
        mol = Chem.MolFromSmiles(i) 
        fpts =  AllChem.GetMorganFingerprintAsBitVect(mol,2,2048)
        mfpts = np.array(fpts)
        Morgan_fpts.append(mfpts)  
    return np.array(Morgan_fpts)
```

```python
Morgan_fingerprints = pd.DataFrame(Morgan_fpts,columns=['Col_{}'.format(i) for i in range(Morgan_fpts.shape[1])])
Morgan_fingerprints
```

```python
mols = [Chem.MolFromSmiles(smiles) for smiles in df['SMILES']]

# Generate Morgan fingerprints for each molecule
Morgan_fpts = [AllChem.GetMorganFingerprintAsBitVect(mol, 2) for mol in mols]

# Convert the fingerprints into a format that can be turned into a DataFrame
Morgan_fpts = [list(fp) for fp in Morgan_fpts]

# Create the DataFrame
Morgan_fingerprints = pd.DataFrame(Morgan_fpts, columns=['Col_{}'.format(i) for i in range(len(Morgan_fpts[0]))])

Morgan_fingerprints
```
